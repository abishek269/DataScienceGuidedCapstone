## 2. Convert the returned JSON object into a Python dictionary.

```python
## pulling and Reading Entire year data
url = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?&start_date=2017-01-01&end_date=2017-12-31&api_key=' + API_KEY
r = requests.get(url)
r_json = r.json()
```

## 3. Calculate what the highest and lowest opening prices were for the stock in this period.

```python
data = r_json['dataset']['data']
```

```python
opening = [value[1] for value in data if value[1] != None]
print("Highest opening price is :", max(opening))
print("Lowest opening price is :", min(opening))
```

## 4. What was the largest change in any one day (based on High and Low price)?

```python
changes = [value[2]-value[3] for value in data if ((value[2] != None) & (value[3] != None))]
print("Largest Change in one day is: ", max(changes))
```

## 5. What was the largest change between any two days (based on Closing Price)?

```python
current_closing = [value[4] for value in data if value[4] != None]
past_closing = [value[4] for value in data if value[4] != None][1:]
diff = [abs(x1 - x2) for (x1, x2) in zip(current_closing, past_closing)]
print("Largest Change between any two days is: ", max(diff))
```

## 6: What was the average daily trading volume during this year?

```python
vol = [value[6] for value in data if value[6] != None]
print("Average daily trading volume is: ", np.mean(vol))
```

## 7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
def find_median(list):
    sorted_list = sorted(list)
    if len(sorted_list)%2 ==1:
        return sorted_list[len(sorted_list)//2]
    else:
        return sum(sorted_list[len(sorted_list)//2-1:len(sorted_list)//2+1])/2.
print("Median trading Vol is: ", find_median(vol))
```
